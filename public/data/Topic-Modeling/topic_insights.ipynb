```python
import pandas as pd

df = pd.read_csv("C:/COOLYEAHH/semester_5/analisis_sosial_media/analystMU_fawza/public/data/Topic-Modeling/document_topics.csv")
display(df.head())
```

```python
# menghitung ada berapa topik
num_topics = df['topic'].nunique()
print(f'Jumlah topik: {num_topics}')

# mencetak semua topik yang ada
topics = df['topic'].unique()
print('Topik yang ada:', topics)
```

```python
# Menampilkan contoh 5 dokumen untuk setiap topik
for topic in topics:
    print(f'\nDokumen untuk Topik {topic}:')
    topic_docs = df[df['topic'] == topic].head(20)
    display(topic_docs)
```

```python
#distribusi topik tabel
topic_summary = df['topic'].value_counts().sort_index().reset_index()
topic_summary.columns = ['topic', 'document_count']
topic_summary['percentage'] = (topic_summary['document_count'] / len(df)) * 100
display(topic_summary)
```

```python
# distribusi topik bar chart
import matplotlib.pyplot as plt
topic_counts = df['topic'].value_counts().sort_index()
plt.bar(topic_counts.index, topic_counts.values)
plt.xlabel('Topik')
plt.ylabel('Jumlah Dokumen')
plt.title('Distribusi Topik dalam Dokumen')
plt.show()
```
